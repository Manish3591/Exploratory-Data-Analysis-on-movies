# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
data.isnull().sum()
```

```python
duplicates = data[data.duplicated()]
duplicates
```

```python
#missing data
total = data.isnull().sum().sort_values(ascending=False)
percent = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
#missing_data.head(20)
print(missing_data)
```

```python
#converting column name to lower case
data.columns= data.columns.str.lower()
data.columns
```

```python
#filling missing metascore values with median
median_meta= data.metascore.median()                                              
data.metascore.fillna(median_meta, inplace = True)
```

```python
#renaming revenue and runtime colums

data.rename(columns={'revenue (millions)':'revenue_millions'},inplace=True)
data.rename(columns={'runtime (minutes)':'runtime_minutes'},inplace=True)
```

```python
#rounding off revenue and metascore values

data['revenue_millions']=data['revenue_millions'].apply(np.floor)
data['metascore']=data['metascore'].apply(np.floor)
```

```python
#converting metascore to dtype int
data['metascore'] = data['metascore'].astype(int)
```

```python
#formatting genre and actors column values

data['genre'] = data['genre'].str.replace(",", "|")
data['actors'] = data['actors'].str.replace(",", "|")
```

```python
#converting row values to lower case where data type is object
for column in data.select_dtypes("object").columns:
    data[column] = data[column].str.lower()
```

```python
#filling missing revenue values with median
median_revenue= data.revenue_millions.median()                                              
data.revenue_millions.fillna(median_revenue, inplace = True)
```

## Highest and lowest earning movies

```python
import pprint
#defining the function
def calculate(column):
    #for highest earned profit
    high= data[column].idxmax()
    high_details=pd.DataFrame(data.loc[high])
    
    #for lowest earned profit
    low= data[column].idxmin()
    low_details=pd.DataFrame(data.loc[low])
    
    #collectin data in one place
    info=pd.concat([high_details, low_details], axis=1)
    
    return info

#calling the function
calculate('revenue_millions')
```

```python
#selecting the movies having profit $50M or more
profit_data = data[data['revenue_millions'] >= 50]

#reindexing new data
profit_data.index = range(len(profit_data))

#we will start from 1 instead of 0
profit_data.index = profit_data.index + 1

#printing the changed dataset
profit_data.head(3)
```

```python
len(profit_data)
```

**429 movies in the dataset have grossed over 50 million dollars**

```python
#function which will take any column as argument from and keep its track 
def profit_genre(column):
    #will take a column, and separate the string by '|'
    profit_genre = profit_data[column].str.cat(sep = '|')
    
    #giving pandas series and storing the values separately
    profit_genre = pd.Series(profit_genre.split('|'))
    
    #arranging in descending order
    count = profit_genre.value_counts(ascending = False)
    
    return count
```

```python
#variable to store the retured value
count = profit_genre('genre')
#printing top 5 values
count.head()
```
